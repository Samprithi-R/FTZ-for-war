```python
from urllib.request import urlopen
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import string
import re
import requests, zipfile, io, os

#Get URL            
url = "http://data.gdeltproject.org/events/index.html"
reqs = requests.get(url)
soup = BeautifulSoup(reqs.text, 'html.parser')
 
urls = []
for link in soup.find_all('a'):
    urls.append(link.get('href'))
    
    
#Getting all the available dates
final_date_list = []
for i in urls:
    da = i.split('.')
    dates = da[0]
    if "2022" in dates:
        try : 
            date_time_obj = datetime.strptime(dates, "%Y%m%d")
            date_time_obj = date_time_obj.strftime("%Y-%m-%d")
            final_date_list.append(date_time_obj)
        except ValueError as ve:
            continue


            
#function to download the data
def download_data(start_date,end_date):
    required_dates = []
    for d in final_date_list:
        if(d>=start_date and d<=end_date):
            required_dates.append(d)


    before = "http://data.gdeltproject.org/events/"
    after = ".export.CSV.zip"
    links = []
    for da in required_dates:
        ss = str(da)
        ss = ss.translate(str.maketrans('', '', string.punctuation))
        ss = before+ss+after
        r = requests.get(ss)
        z = zipfile.ZipFile(io.BytesIO(r.content))
        to_path = "C:\\Users\\SAMPRITHI\\Desktop\\sample_downloads"
        x = z.extractall(to_path)
        for filename in z.namelist():
            os.rename(os.path.join(to_path, filename), os.path.join(to_path, filename.replace(".export", "")))
            
            

download_data('2022-02-28','2022-09-11')

```
