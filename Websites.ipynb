```python
from os import listdir

def find_csv_filenames( path_to_dir, suffix=".CSV" ):
    filenames = listdir(path_to_dir) 
    return [ filename for filename in filenames if filename.endswith( suffix ) ]



fl = find_csv_filenames('C:\\Users\\SAMPRITHI\\Desktop\\sample_downloads')
```

```python
import pandas as pd
```

```python
websites = []
for fi_name in fl:
    df = pd.read_csv(r'C:\Users\SAMPRITHI\Desktop\sample_downloads\{0}'.format(fi_name), delimiter='\t')
    def get_org(url):
        return url.split('/')[2]
    orgs = []
    for row in df.itertuples():
        orgs.append(get_org(row._58))
    df['org'] = orgs   
    aa = df.loc[df.org.isin(['www.rt.com', 'sputniknews.com'])][df.columns[-2]]
    final_web = aa.unique()
    websites.extend(final_web)
    
print(len(websites))
#print(websites)
    
```

```python
websites = pd.DataFrame(websites,columns = ['websites'])
websites.to_csv('C:\\Users\\SAMPRITHI\\Desktop\\Research\\websites.CSV') 
```
